```python
#loading necessary packages
from sqlalchemy import create_engine
import numpy as np
import pandas as pd
```

```python
# Create engine: engine
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

```python
#read tables names
table_names = engine.table_names()
table_names
```

```python
pd.read_sql_query('SELECT * FROM Facilities', engine).head()
```

```python
#question 10
query_Q10 = """SELECT
`facility name`,
SUM(`cost`) AS "revenue"
FROM
Members AS M
INNER JOIN (
	SELECT
	B.memid,
	F.name AS "facility name",
	CASE 
		WHEN B.memid = 0 THEN B.slots * F.guestcost
		ELSE B.slots * F.membercost
	END AS "cost"
	FROM
	Bookings AS B
	INNER JOIN Facilities AS F ON F.facid = B.facid
) bookings_all ON bookings_all.memid = M.memid
GROUP BY `facility name`
HAVING SUM(`cost`) < 1000
ORDER BY `revenue`"""
pd.read_sql_query(query_Q10, engine)
```

```python
#question 11
query_Q11 = """
SELECT
M.surname || " " || M.firstname  AS "member name",
CASE
	WHEN R.memid > 0 THEN R.surname || " " || R.firstname 
	ELSE 'None'
END AS "recommended by"
FROM
Members AS M
LEFT JOIN Members AS R ON R.memid = M.recommendedby
ORDER BY M.surname, M.firstname
"""
pd.read_sql_query(query_Q11, engine)
```

```python
#question 12
query_Q12 = """
SELECT
F.name AS `facility name`,
MAX(M.firstname || " " || M.surname) AS `member name`,
COUNT(DISTINCT B.bookid) AS `count of bookings`
FROM
Members AS M
INNER JOIN Bookings AS B ON B.memid = M.memid
INNER JOIN Facilities AS F ON F.facid = B.facid
WHERE
M.memid > 0
GROUP BY F.name, M.memid
"""
pd.read_sql_query(query_Q12, engine)
```

```python
#question 13
query_Q13 = """
SELECT
F.name AS `facility name`,
strftime('%m', B.starttime) AS `month`,
COUNT(DISTINCT B.bookid) AS `count of bookings`
FROM
Members AS M
INNER JOIN Bookings AS B ON B.memid = M.memid
INNER JOIN Facilities AS F ON F.facid = B.facid
WHERE
M.memid > 0
GROUP BY F.name, strftime('%m', B.starttime)
"""
pd.read_sql_query(query_Q13, engine)
```
